```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(11)

X_low_var = np.random.normal(1,0.05,1000)

Y_class = np.random.choice([0,1], size=1000, p=[0.5,0.5])

df = pd.DataFrame({
    'Cecha_X': X_low_var,
    'Klasa_Y': Y_class
})
sum(Y_class==1)
```

```python
plt.figure(figsize=(10, 6))

# Rysowanie histogramu z podziałem na klasy (hue='Klasa_Y')
sns.histplot(data=df, x='Cecha_X', hue='Klasa_Y', alpha=0.6, multiple="stack")

plt.title('Dystrybucja Cecha_X dla Klasy 0 i Klasy 1')
plt.xlabel('Wartość Cecha_X')
plt.ylabel('Liczba obserwacji')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()
X = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
y = cancer_data.target # 0 - nowotwór złośliwy, 1 - łagodny
```

```python
Var = X.var()
var_sorted=Var.sort_values(ascending=False)

print(var_sorted.head(5))
print(var_sorted.tail(7))
```

```python
df = X.copy()
df['Klasa']=y
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, 
             x='mean smoothness', 
             hue='Klasa', 
             alpha=0.6, 
             multiple="stack",
             palette=["#da1b1b", "#4995cb"])

plt.title('Histogram zmiennej mean smoothness w podziale na klasy')
plt.xlabel('Wartość cechy')
plt.ylabel('Liczba obserwacji')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Rysowanie histogramu z podziałem na klasy (hue='Klasa_Y')
sns.histplot(data=df, 
             x='worst area', 
             hue='Klasa', 
             alpha=0.45, 
             bins=30,
             kde=True,                 # Włącza krzywą gęstości
             stat="density",           # Ustawia oś Y na gęstość prawdopodobieństwa
             common_norm=False, 
             palette=["#da1b1b", "#2485cb"],
             line_kws={'linewidth': 2, 'alpha': 1})

plt.title('Dystrybucja zmiennej worst area w podziale na klasy')
plt.xlabel('Wartość cechy')
plt.ylabel('Gęstość')
plt.grid(True, alpha=0.3)
plt.show()
```
