```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()
X = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
y = cancer_data.target # 0 - nowotwór złośliwy, 1 - łagodny
```

```python
X.describe()
```

```python
Var = X.var()
var_sorted=Var.sort_values(ascending=False)

print(var_sorted.head(5))
print(var_sorted.tail(7))
```

```python
df = X.copy()
df['Klasa']=y
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, 
             x='mean smoothness', 
             hue='Klasa', 
             alpha=0.45, 
             bins=30,
             kde=True,                 
             stat="density",           
             common_norm=False, 
             palette=["#da1b1b", "#2485cb"],
             line_kws={'linewidth': 2, 'alpha': 1})

plt.title('Dystrybucja zmiennej mean smoothness w podziale na klasy')
plt.xlabel('Wartość cechy')
plt.ylabel('Gęstość')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, 
             x='worst area', 
             hue='Klasa', 
             alpha=0.45, 
             bins=30,
             kde=True,                 
             stat="density",           
             common_norm=False, 
             palette=["#da1b1b", "#2485cb"],
             line_kws={'linewidth': 2, 'alpha': 1})

plt.title('Dystrybucja zmiennej worst area w podziale na klasy')
plt.xlabel('Wartość cechy')
plt.ylabel('Gęstość')
plt.grid(True, alpha=0.3)
plt.show()
```
