```python
from sklearn.datasets import make_classification
import pandas as pd
import numpy as np
```

```python
X, y = make_classification(n_samples=100,
                           n_features=20,
                           n_informative=10,
                           n_redundant=5,
                           n_repeated=1,
                           n_classes=2, 
                           random_state=111,
                           shuffle=False)

df=pd.DataFrame(X)
df['Klasa'] = y
```

* 0-9 cechy informatywne
* 10-14 kombinacje liniowe innych cech
* 15 cecha powtorzona
* 16-19 szum
